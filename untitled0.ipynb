```python
try:
  x = int(input("Enter the first  number: "))
  y = int(input("Enter the second  number: "))
  result = x + y
  print(f"User Input Addition: {x} + {y} = {result}")
except ValueError:
  print("Please enter valid integers!")
```

```python
with open("multiple_input.txt","w") as f:
  f.write("3\n5\n7")
  try:
    with open("multiple_input.txt","r") as f:
      numbers = [int(line.strip()) for line in f.readlines()]
      result = sum(numbers)
      print(f"Multiple Input Read from File: {numbers} = {result}")
  except FileNotFoundError:
    print("File not found!")
  except ValueError:
    print("Error in reading or converting the input!")
```
